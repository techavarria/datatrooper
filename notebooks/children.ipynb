```python
import pandas as pd
import matplotlib as mpl
import seaborn as sns
import matplotlib.pyplot as plt
import pycountry
from pprint import pprint
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import plotly.graph_objs as go
import plotly
import numpy as np


sns.set_context("talk")
plt.style.use('ggplot')
mpl.rcParams['axes.labelpad']='20'
```

```python
df = pd.read_csv("../datos/unicef_data.csv")
```

```python
columns = ["Geographic Area", "Indicator", "Sex", "Time Period", "Observation Value"]
df = df.loc[:,columns]
js, js_1a4, js_5a14, js_infant, js_neonatal = {}, {}, {}, {}, {}
sexo = "Total"
for i, year in enumerate(df["Time Period"]):
    try:
        js[str(year)].append(df.loc[i,"Observation Value"])
    except Exception:
        js[str(year)] = [df.loc[i,"Observation Value"]]
    if (df.loc[i,"Indicator"] == "Child deaths age 1 to 4") and (df.loc[i,"Sex"] == sexo):
        try:
            js_1a4[str(year)].append(float(df.loc[i,"Observation Value"]))
        except Exception:
            js_1a4[str(year)] = [float(df.loc[i,"Observation Value"])]
    if (df.loc[i,"Indicator"] == "Deaths age 5 to 14") and (df.loc[i,"Sex"] == sexo):
        try:
            js_5a14[str(year)].append(float(df.loc[i,"Observation Value"]))
        except Exception:
            js_5a14[str(year)] = [float(df.loc[i,"Observation Value"])]
    if (df.loc[i,"Indicator"] == "Infant deaths") and (df.loc[i,"Sex"] == sexo):
        try:
            js_infant[str(year)].append(float(df.loc[i,"Observation Value"]))
        except Exception:
            js_infant[str(year)] = [float(df.loc[i,"Observation Value"])]
    if (df.loc[i,"Indicator"] == "Neonatal deaths") and (df.loc[i,"Sex"] == sexo):
        try:
            js_neonatal[str(year)].append(float(df.loc[i,"Observation Value"]))
        except Exception:
            js_neonatal[str(year)] = [float(df.loc[i,"Observation Value"])]
            

pprint(js_1a4)
```

```python
titulo = "Mortalidad infantil"

for i, year in enumerate(list(js_1a4.keys())):
    y = []
    
    try:
        y.append(np.sum(js_neonatal[year]))
    except Exception:
        y.append(0)
    try:
        y.append(np.sum(js_1a4[year]))
    except Exception:
        y.append(0)
    try:
        y.append(np.sum(js_5a14[year]))
    except Exception:
        y.append(0)

        
    x = ["neonatales", "1 a 4 años","5 a 14 años"]
    
    fig = go.Figure(
    data=[go.Bar(x=x,y=y)],
    layout_title_text= titulo + ' ' + year[0:4]
    )
    fig.update_layout(
        title="Mortalidad infantil en " + year[0:4],
        font=dict(
            family="Courier New, monospace",
            size=28,
            color="#7f7f7f"
        )
    )
    fig.update_yaxes(range=[0, 5500000])

    plotly.offline.plot(fig, filename=f'images/newplot({year[0:4]}).html', image = 'png', image_filename=f'image({year[0:4]})',auto_open=True)
```
